```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://stepik.org/media/attachments/course/4852/train_data_tree.csv')
```

```python
data.head()
```

```python
X = data.drop('num', axis=1)
y = data['num']
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
result = cross_val_score(clf, X, y)
result
```

```python
result.mean()
```

```python
clf.fit(X, y)
```

```python
plt.figure(figsize=(6, 5))
plot_tree(clf, filled=True);
```

```python
plt.figure(figsize=(6, 5), dpi=100)
plot_tree(clf, fontsize=8, filled=True, feature_names=list(X));
```

```python
l_node = clf.tree_.children_left[0]
r_node = clf.tree_.children_right[0]
```
