```python
import pandas as pd
import numpy as np
from sklearn.tree import plot_tree, DecisionTreeClassifier
```

```python
data = pd.read_csv('https://stepik.org/media/attachments/course/4852/cats.csv', index_col=0)
```

```python
data
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
X = data[['Шерстист']]
y = data['Вид']
```

```python
clf.fit(X, y)
```

### Шерстист

```python
plot_tree(clf, filled=True);
```

```python
X = data[['Гавкает']]
y = data['Вид']
```

### Гавкает

```python
plot_tree(clf.fit(X, y), filled=True);
```

```python
X = data[['Лазает по деревьям']]
y = data['Вид']
```

### Лазает по деревьям
