```python
students_performance.groupby('gender').mean()
```

### А здесь опять все хорошо.

```python
students_performance.groupby('gender').aggregate({'math score': 'mean', 'reading score': 'mean'})
```

```python
students_performance.groupby('gender', as_index=False)\
                    .aggregate({'math score': 'mean', 'reading score': 'mean'})\
                    .rename(columns={'math score': 'mean_math_score', 'reading score': 'mean_reading_score'})
```

```python
students_performance.groupby('gender', as_index=False)\
                    .aggregate({'math score': ['mean', 'count', 'std'], 'reading score': ['std', 'min', 'max']})
```

```python
students_performance.groupby('gender', as_index=False)\
                    .aggregate(mean_math_score = ('math score', 'mean'), mean_reading_score = ('reading score', 'mean'))
```

```python
students_performance.groupby('gender').aggregate({'math score': 'mean'})
```

```python
students_performance.groupby('gender').agg({'math score': 'mean', 'reading score': 'mean', 'writing score': 'mean'})
```

```python
students_performance.groupby('gender')[['math score', 'reading score', 'writing score']].agg(['mean', 'std', 'median', 'min', 'max'])
```
