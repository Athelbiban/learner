```python
import pandas as pd
```

```python
concentrations = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
```

```python
concentrations
```

```python
concentrations.pivot_table(values=['sucrose', 'alanin', 'citrate', 'glucose', 'oleic_acid'], columns='genus', aggfunc='mean')
```

```python
mean_concentrations = concentrations.groupby('genus')[['sucrose', 'alanin', 'citrate', 'glucose', 'oleic_acid']].mean()
```
