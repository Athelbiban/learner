```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
students_performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
students_performance.head()
```

```python
students_performance['math score'].hist()
```

```python
students_performance.plot.scatter(x='math score', y='reading score');
```

```python
sns.lmplot(x='math score', y='reading score', data=students_performance);
```

```python
sns.lmplot(x='math score', y='reading score', hue='gender', data=students_performance);
```

```python
sns.lmplot(x='math score', y='reading score', hue='gender', data=students_performance, fit_reg=False);
```

```python
ax = sns.lmplot(x='math score', y='reading score', hue='gender', data=students_performance, fit_reg=False);
ax.set_xlabels('')
ax.set_ylabels('')
```
