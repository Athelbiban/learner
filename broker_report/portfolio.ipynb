```python
benchmark = {
    'SBMX': 10,
    'RU000A101FA1': 20,
    'SBER': 15,
    'MTSS': 10,
    'GAZP': 8,
    'LKOH': 10,
    'MGNT': 9,
    'YNDX': 3,
    'CHMF': 5,
    'AFKS': 5,
    'IRAO': 5
}
```

```python
# portfolio.loc['RU000A101FA1', 'Котировки'] = portfolio.loc['RU000A101FA1', 'Котировки'] / 100 * 1000
portfolio['Текущая цена'] = portfolio['Котировки'] * portfolio['Количество']
portfolio['P/L'] = portfolio['Текущая цена'] - portfolio['Средняя цена'] * portfolio['Количество']
portfolio['P/L, %'] = (portfolio['Котировки'] * 100 / portfolio['Средняя цена'] - 100).round(2)
current_price_sum = (portfolio['Текущая цена'].sum()).round(2)
average_price_sum = ((portfolio['Средняя цена'] * portfolio['Количество']).sum()).round(2)
PL_sum = (portfolio['P/L'].sum()).round(2)
PL_per_sum = current_price_sum * 100 / average_price_sum - 100
portfolio['Пропорции, %'] = (portfolio['Текущая цена'] / current_price_sum * 100).round(2)
proportions_sum = (portfolio['Пропорции, %'].sum()).round(2)
portfolio['Бенчмарк, %'] = benchmark
```
