```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('training_log.csv')
```

перевел данные столбца `date` в формат `datetime`

```python
df['date'] = pd.to_datetime(df['date'])
```

общие данные по фрейму

```python
df.describe()
```

данные сгруппированные по видам тренировок

```python
df.groupby('train')[['speed', 'error']].describe()
```
